```python
import random
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
from collections import Counter
```

```python
# Load data
nobel_laureates_counterparts_pubs = pd.read_csv('MAG-nobel/2 Nobel laureates counterparts/counterparts pubs.zip')
nobel_laureates_counterparts = pd.read_csv('MAG-nobel/2 Nobel laureates counterparts/counterparts.zip')
nobel_laureates = pd.read_csv('MAG-nobel/Nobel laureates.csv')
nobel_laureates_pubs = pd.read_csv('MAG-nobel/Nobel Laureates pubs.zip')
MAG_hit = pd.read_csv('MAG-all/hitlst.zip')
MAG_hit['Is Hit'] = 1
```

```python
# Fill missing values
nobel_laureates_pubs['Is Hit'] = nobel_laureates_pubs['Paper ID'].map(MAG_hit.set_index('0')['Is Hit'])
nobel_laureates_pubs.fillna(0, inplace=True)

nobel_laureates_counterparts_pubs['Is Hit'] = nobel_laureates_counterparts_pubs['paperid'].map(MAG_hit.set_index('0')['Is Hit'])
nobel_laureates_counterparts_pubs.fillna(0, inplace=True)
```

# Career

```python
# Career
bohr_pubs = nobel_laureates_pubs[nobel_laureates_pubs['Laureate name'] == 'bohr, a.']
```

```python
x1 = bohr_pubs['Pub year'].values
y1 = bohr_pubs['paper cit'].values
x1h = bohr_pubs[bohr_pubs['Is Hit'] == 1]['Pub year'].values
y1h = bohr_pubs[bohr_pubs['Is Hit'] == 1]['paper cit'].values
x1nobel = bohr_pubs[bohr_pubs['Is prize-winning paper'] == 'YES']['Pub year'].values
y1nobel = bohr_pubs[bohr_pubs['Is prize-winning paper'] == 'YES']['paper cit'].values
```

```python
# Plot
plt.rcParams['font.sans-serif'] = ['Arial']
f, ax = plt.subplots(figsize=(15, 10))
ax.stem(x1, y1, linefmt='C0-.', markerfmt='C0o', basefmt=' ', label='Other pubs')
ax.stem(x1h, y1h, linefmt='C0-.', markerfmt='C3*', basefmt=' ')
ax.scatter(x1h, y1h, s=680, marker='*', color='red', label='Highly cited pubs')
ax.scatter(x1nobel, y1nobel, s=4000, marker='*', color='#F1C40F', label='Nobel Prize pubs')
ax.set_yscale('log')
plt.xlim(1948, 1980)
plt.xticks([1950, 1960, 1970, 1980], size=48)
plt.yticks([10e0, 10e1, 10e2, 10e3], size=48)
plt.legend(fontsize=35)
#plt.ylabel(r'impact', fontsize=52)
plt.savefig(r"F:\A J YANG PAPER\Nobel analysis\PNAS representative evaluation\figs\fig 1\a1.pdf",bbox_inches = 'tight')
```
